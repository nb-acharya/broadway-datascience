## handling null (NaN) values
1. Fill data
   - mean  (numeric record) -> if data is in range fill with mean value
   - median  (numeric record) -> if data is not in range fill with median value
   - mode  (categorical record)
   - custom record  (both numeric and categorical)
3. Drop Row/Column

```python
commission_mean = round(df['COMMISSION_PCT'].mean(),2)
# df['COMMISSION_PCT'].fillna(commission_mean, inplace = True)
```

```python
df['COMMISSION_PCT'] = df['COMMISSION_PCT'].fillna(commission_mean)
print('commission pct is filled with mean value')
```

```python
df['COMMISSION_PCT'].median()
```

```python
df
```

```python
manager_mode = df['MANAGER_ID'].mode()[0]
manager_mode
```

```python
department_mode = df['DEPARTMENT_ID'].mode()[0]
department_mode
```

```python
df["MANAGER_ID"]=df["MANAGER_ID"].fillna(manager_mode)
df['DEPARTMENT_ID'] = df['DEPARTMENT_ID'].fillna(department_mode)
```

### Drop data

```python
df = pd.read_csv("Dataset/Employees.csv")
df
```

```python
# default -> row wise delete
#df.dropna(inplace=True) # doesnt remove in original df, so use inplace=True
df = df.dropna(axis=1) # you can use this too
df
```

#### parameters of dropna()
- (axis = 0 (default)) -> Row and (axis = 1) -> Column
- how = any(default)/all
- inplace = False(default)/True
