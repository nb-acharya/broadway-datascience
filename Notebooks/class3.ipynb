```python
#find details of employees whose salary is more than 18000
filtered_salary=df['SALARY'] > 18000
filtered_salary
```

```python
df[filtered_salary]
```

```python
#find details of employees whose salary is in the range of 10000 and 20000, using and , between

sal_range = (df["SALARY"] >= 10000) & (df['SALARY'] <= 20000)
sal_range

sal_between = df["SALARY"].between(10000,20000)
sal_between

#if dataset is small, then using and between doesnt matter
#if dataset is large, then use and, cause between is slower 
# reasoning: and doesnt have to perform the operation but the between has to perform the operation for one condition and again for another condition
```

```python
df[sal_range]
df[sal_between]
```

```python
#categorical data: use OR ok !!! ;)
#find details of employees who works in job SA_REP, IT_PROG

job_details = (df['JOB_ID'] == 'SA_REP') | (df['JOB_ID'] == 'IT_PROG')
df[job_details]
```

```python
#categorical data: use OR ok !!! ;)
#find details of employees who works in job SA_REP, IT_PROG using isin

job_det = df['JOB_ID'].isin(['SA_REP','IT_PROG']) # if we are calling the function, we have to use list in the parameters if we have multiple values to pass
df[job_det]
```

```python
# find details of employees whose salary is in range of 6000 to 10000 and they must be working on job SA_REP

# using and condition
emp_detail1 = (df["SALARY"] >= 6000) & (df['SALARY'] <= 10000) & (df['JOB_ID'] == "SA_REP")
df[emp_detail1]
```

```python
# using between and isin
emp_detail2 = df["SALARY"].between(6000,10000) & df['JOB_ID'].isin(['SA_REP'])
df[emp_detail2]
```
